##**Matplotlib**

```python
import matplotlib.pyplot as plt
```

###we can perform diffrent types of anylasis
**It has 3 types:**
1. Univariate : single feature
2. Bi-Variate : 2 features
3. Multi-Variate : more than 2 features (Sate vs Population -->gender)

```python
days=['monday','tuesday','wednesday','thursday','friday']
rain=[120,30,25,20,80]
```

```python
#Bi- variate plots - line ,scatter, bar ,barh
plt.plot(days,rain,'m-.h',label='rainfall in mm')
plt.legend()
plt.grid()
plt.xlabel('Days')
plt.ylabel('Rainfall in mm')
plt.title('Rain Analysis')
plt.savefig('rain.png',dpi=400)
```

```python
#bar plot
plt.bar(days,rain,color= ['r','y','m','c','k'],label='Daywise rainfall in mm')
plt.grid()
plt.legend()
plt.xlabel('Days')
plt.ylabel('Rainfall in mm')
plt.title('Rain Analysis')
```

```python
#bar plot
plt.barh(days,rain,color= ['r','y','m','c','k'],label='Daywise rainfall in mm')
plt.grid()
plt.legend()
plt.ylabel('Days')
plt.xlabel('Rainfall in mm')
plt.title('Rain Analysis')
```

###**Seaborn**

```python
import seaborn as sns
```

```python
#load dataset
sns.get_dataset_names()
```

```python
p = sns.load_dataset('penguins')
p.head()
```

```python
p['species'].unique()
```

```python
#using matplotlib
p['species'].value_counts().plot(kind='bar')
```

**using seaborn**

```python
sns.countplot(data=p,x='species')
```

```python
sns.countplot(data=p,x='island')
```

```python
sns.countplot(data=p,x='species',hue='sex',palette='pastel')
```

```python
sns.stripplot(data=p,x='species',y='body_mass_g',hue='island')
```
