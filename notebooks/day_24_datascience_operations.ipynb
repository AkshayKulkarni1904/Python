## Setup

```python
path = '/content/CO2 Emissions_Canada.csv'
```

### Data Loading

The first step in any data analysis workflow is to load your data into a suitable structure. Here, we load a CSV file into a pandas DataFrame, which is a powerful and flexible two-dimensional tabular data structure.

```python
#Load data set
import pandas as pd

c=pd.read_csv(path)
c.head(2)
```

#### Overview of DataFrame

After loading, it's good practice to get a quick overview of the DataFrame, such as its columns and the first few rows, to understand its structure and content.

```python
c.columns
```

```python
c.columns
```

###**Filter based selection / Conditional selection**

## Data Filtering

Data filtering allows you to select specific rows or columns based on certain conditions. This is a crucial step for focusing on relevant subsets of your data.

```python
c[:2]
```

### Basic Slicing

Slicing with integer indexing is a way to select a range of rows. For example, `c[:2]` selects the first two rows.

```python
c[:2]
```

```python
c['Make'].unique()
```

### Using `unique()` to inspect column values

The `unique()` method is useful for understanding the distinct values present in a specific column, which can be helpful for filtering or categorical analysis.

```python
#Select records of BMW
#syntax : c.query("expression")
c.query("Make=='BMW'")
```

### Filtering with `query()` method

The `query()` method provides a string-based syntax to filter DataFrame rows, often making complex selections more readable. It's particularly convenient for expressions involving column names with spaces, which can be enclosed in backticks.

```python
c.query("Make=='BMW' and Cylinders>6")
```

#### Multiple Conditions with `query()`

You can combine multiple conditions within the `query()` method using logical operators like `and`, `or`, and `not`.

```python
c.columns
```

```python
display(c.columns)
```

```python
# use ` ` to fetch data with space or special characters from table
c.query("Make=='BMW' and `Engine Size(L)`>6")
```

### Filtering with Boolean Indexing

Boolean indexing uses a boolean Series (True/False values) to select rows from a DataFrame where the corresponding boolean value is `True`. This is a powerful and flexible way to filter data.

```python
#Access/Filter data using Indexing
c['Vehicle Class']=='FULL-SIZE'
```

#### Single Condition Boolean Indexing

To filter using a single condition, you create a boolean Series from a column and then pass it to the DataFrame.

```python
#Using boolean indexing
c[c['Vehicle Class']=='FULL-SIZE']
```

#### Multiple Conditions Boolean Indexing

For multiple conditions, you combine boolean Series using bitwise logical operators: `&` (and), `|` (or), and `~` (not). Remember to enclose each condition in parentheses.

```python
c['Make']=='ACURA'
```

```python
c['Make']=='ACURA'
```

```python
c[c['Make']=='ACURA']
```

#### Example: Filtering for 'ACURA' Make

```python
#boolean value
(c['Make']=='ACURA') & (c['Cylinders']==6)
```

#### Example: Filtering for 'ACURA' Make with 6 Cylinders

```python
#Required output
# use bitwise & for multiple comparisons
c[(c['Make']=='ACURA') & (c['Cylinders']==6)]
```

## Groupby Operations

The `groupby()` method is used for grouping data based on one or more columns and then applying an aggregation function to the groups. This allows for powerful summarization and analysis of data.

##**Groupby operations**

```python
list(c.groupby(['Make']))
```

### Listing Groups

When you group a DataFrame, you can inspect the individual groups that are formed. The output below shows the groups as (key, DataFrame) pairs.

```python
#fetch a particuler group
c.groupby(['Make']).get_group('TOYOTA')
```

```python
# Corrected version to avoid FutureWarning in get_group
c.groupby(['Make']).get_group(('TOYOTA',))
```

### Fetching a Specific Group

After grouping, you can select a particular group using `get_group()`. Note the `FutureWarning` in the original notebook, which suggests passing a tuple `('TOYOTA',)` to `get_group` for single-column groupings in future pandas versions. This is shown in the corrected version below to avoid the warning.

```python
c.columns
```

```python
import numpy as np
import warnings
#to remove warnnings if we have multiple columns
warnings.filterwarnings('ignore')

cols = [ 'Fuel Consumption City (L/100 km)',
       'Fuel Consumption Hwy (L/100 km)', 'Fuel Consumption Comb (L/100 km)',
       'Fuel Consumption Comb (mpg)', 'CO2 Emissions(g/km)']
c.groupby(['Make']).get_group('TOYOTA').loc[:,cols].agg([min,max,np.mean])
```

### Aggregating Grouped Data

After grouping, you can apply various aggregation functions (like `min`, `max`, `mean`, `sum`, `count`) to the groups. This provides summary statistics for each group.
