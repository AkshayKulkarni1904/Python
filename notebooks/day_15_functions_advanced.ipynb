### Understanding Function Arguments

Python functions offer flexible ways to pass data to them using arguments. Here's a brief breakdown of the types mentioned:

*   **Positional Arguments:** Arguments passed in the order they are defined in the function signature. Their position matters.
*   **Keyword Arguments:** Arguments passed by explicitly naming the parameter (e.g., `func(param_name=value)`). Their order does not matter.
*   **Default Arguments:** Parameters that have a default value. If a value is not provided during the function call, the default value is used.
*   **Variable-Length Arguments (`*args` and `**kwargs`):**
    *   `*args`: Allows a function to accept an arbitrary number of positional arguments. These arguments are collected into a tuple.
    *   `**kwargs`: Allows a function to accept an arbitrary number of keyword arguments. These arguments are collected into a dictionary.

### The `return` Keyword

The `return` statement is used in a function to send a result back to the caller. When `return` is executed, the function immediately stops, and the value (or `None` if no value is specified) is sent back. This allows functions to produce outputs that can be used in other parts of the program.

```python
#Function will work on specific rquirments
#it will generate one result
#this result we need to use as an input to the outside of the function
#for this purpose we need to use return keyword
def test(balance):
  if balance <=0:
    return'No'
  else:
    return'Yes'

result=test(-100)

if result=='Yes':
  print('You can perform transcation')
else:
  print('You can not perform transcation')
```

The `test` function above demonstrates how `return` is used to pass a string (`'No'` or `'Yes'`) back to the `result` variable. This returned value then dictates the flow of the `if/else` statement outside the function, enabling decision-making based on the function's outcome.

```python
#purchase and billing example using return
items ={'rice':56,'chana':30,'wheat':60,'parle_g':10,'milk':35}
bill=[]
def purchase(*args):
  for item in args:
    bill.append(items[item])
  return bill
def check(total):
  return sum(total)
#purchase calling
total = purchase('wheat','milk')
print('items amount',total)

#calling check
ans=check(total)
print('Total amount',ans)
```

### Chained Functions and Global State Modification

This example showcases how functions can interact. The `purchase` function takes variable positional arguments (`*args`) to process items, calculating their prices and appending them to a global `bill` list. The `check` function then takes this list (or any list of numbers) and returns its sum.

**Note on Global Variables:** While modifying global variables directly (like `bill` here) can be convenient, it's generally good practice to minimize this to avoid unexpected side effects and make your code easier to debug and reuse. It's often better to pass variables as arguments or return new values.

```python
#calculate % and grade
def calculate_per(marks_obtained,total_marks):
  percentage=(marks_obtained/total_marks)*100
  return percentage
def grade(percentage):
  if percentage>=90:
    return'A'
  elif percentage>=80:
    return'B'
  elif percentage>=70:
    return'C'
  elif percentage>=55:
    return'D'
  elif percentage>=40:
    return'E'
  else:
    return'F'

per = calculate_per(300,500)
result = grade(per)
print('Percentage',per)
print('Grade',result)
# print('Total percentage',calculate_per(300,500))
# print('Grade:',grade(calculate_per(20,100)))
```

### Function Composition and Reusability

Here, two functions, `calculate_per` and `grade`, work together. `calculate_per` computes a percentage based on marks, and `grade` then uses this percentage to determine a letter grade. This demonstrates **function composition**, where the output of one function (`calculate_per`) becomes the input for another (`grade`), allowing for modular and reusable code.

#### Question 2: Modifying Functions

To include quantity, the `purchase` function would need to accept items and their quantities. The `bill` list would store the total cost for each item (price * quantity), and the `check` function would remain the same as it sums the `bill`.

Here's how you could modify the functions:

```python
items ={'rice':56,'chana':30,'wheat':60,'parle_g':10,'milk':35}
bill=[]

def purchase_with_quantity(*args):
  for i in range(0, len(args), 2):
    item = args[i]
    quantity = args[i+1]
    bill.append(items[item] * quantity)
  return bill

def check(total):
  return sum(total)

# purchase calling with quantity
# Clear bill first to avoid cumulative effect from previous runs
bill = []
total_with_quantity = purchase_with_quantity('wheat', 2, 'milk', 1)
print('Items and their total costs:', total_with_quantity)

# calling check
ans_with_quantity = check(total_with_quantity)
print('Total amount with quantity:', ans_with_quantity)
```

#### Question 7: Scope and Side Effects (Conceptual)

In the `purchase` function, directly modifying the global `bill` list is an example of a **side effect**.

**Pros of modifying a global list directly:**
*   **Simplicity for small scripts:** In very small, self-contained scripts, it might seem simpler to directly update a global variable without passing it around explicitly.

**Cons of modifying a global list directly:**
*   **Reduced Readability and Maintainability:** It becomes harder to understand which parts of the code are modifying the `bill` list. If the list is changed unexpectedly, it's difficult to trace the source of the change.
*   **Increased Coupling:** The function becomes tightly coupled to the global `bill` variable. If you want to reuse the `purchase` logic with a different bill (e.g., for another customer), you would have to manage global state carefully or refactor the function.
*   **Difficult to Test:** Testing functions with side effects is harder because you need to manage the global state before and after each test.
*   **Potential for Bugs:** Unintended modifications to shared global state are a common source of bugs in larger programs.

**Alternative Approach: Returning the modified data (without side effects)**

A better approach is for the function to return the data it processes or creates, rather than modifying global state. For example, `purchase` could return the list of calculated item costs.

Here's how `purchase` could be rewritten to avoid modifying a global `bill` directly:

```python
items ={'rice':56,'chana':30,'wheat':60,'parle_g':10,'milk':35}

def purchase_no_side_effect(*args):
  local_bill_items = [] # Create a local list
  for item in args:
    local_bill_items.append(items[item])
  return local_bill_items # Return the new list

def check(total):
  return sum(total)

# Now, when calling purchase, you assign its return value
my_customer_bill_items = purchase_no_side_effect('wheat', 'milk')
print('Customer 1 items amount:', my_customer_bill_items)
print('Customer 1 total amount:', check(my_customer_bill_items))

# You can easily have another customer's bill without interference
another_customer_bill_items = purchase_no_side_effect('rice', 'chana', 'parle_g')
print('Customer 2 items amount:', another_customer_bill_items)
print('Customer 2 total amount:', check(another_customer_bill_items))
```
