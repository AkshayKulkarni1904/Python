###Interview question

####Diffrence between iloc and loc?

*`iloc`* is index based where stop is exclusive and *`loc`* is lable based where stop is inclusive

- *`iloc` (integer location)*: Selects data by row and column positions (integer indices). It does not accept labels or boolean arrays and uses exclusive slicing for ranges.
- *`loc` (label location)*: Selects data by row and column labels. It can accept labels, boolean arrays, or slices, and uses inclusive slicing for ranges.

```python
#How to add a row in df
new_player = pd.DataFrame({
    "ID":[99998],
    "Name":["Rahul"],
    "Age":[21],
    "Nationality":["India"],
    "Overall":[72],
    "Potential":[80],
    "Club":[" FC"],
    "Value":["80000"],
    "Wage":["3000"],
    "Preferred Foot":["Left"],
    "International Reputation":[1],
    "Skill Moves":[2],
    "Position":["CM"],
    "Joined":["2024"],
    "Contract Valid Until":["2028-01-01"],
    "Height":["175 cm"],
    "Weight":["70 kg"],
    "Release Clause":["1000000"]
})

df = pd.concat([df, new_player], ignore_index=True)

df.tail()
```

```python
#Append Two DataFrames
#Concatenate Two or More DataFrames
df2= df.tail(5)
df2
```

```python
new_df=pd.concat([df,df2],ignore_index=True)
new_df.tail()
```

```python
#Vertical Concatenation (Rows)
result=pd.concat([df,df2])
result.tail()
```

```python
#Horizontal Concatenation (Columns)
leaft = df[["Name",'Age']].head()
right = df[["Overall","Club"]].head()
result=pd.concat([leaft,right],axis=1)
result
```

```python
#Concatenate Three DataFrames
df1 = df.head(2)
df2 = df.iloc[2:4]
df3 = df.tail(2)
df_concat = pd.concat([df1, df2, df3], ignore_index=True)
df_concat
```

```python
club = pd.DataFrame({
    "Club":["Juventus","Real Madrid","FC Barcelona"],
    "Country":["Italy","Spain","Spain"]
})

club
```

```python
#Merge using the common column Club
merged = pd.merge(df,club,on='Club')
merged.head()
```

```python
#Left Merge
pd.merge(df,club,on='Club',how='left')
```

#

```python
#Right Merge
pd.merge(df,club,on='Club',how='right')
```

```python
#Outer Merge
pd.merge(df,club,on='Club',how='outer')
```

```python
#Join Operation
#Create two small DataFrames.
country = pd.DataFrame({
    "Country":["India","Italy","Spain","Germany"],
    "Capital":["New Delhi","Rome","Madrid","Berlin"]
})
country.index = [0,1,2,3]
country
```

```python
left = df[["Name","Nationality"]].head(3)

right = pd.DataFrame({
    "Nationality":["Portugal","Argentina","Brazil"],
    "Continent":["Europe","South America","South America"]
}).set_index("Nationality")

left
```
