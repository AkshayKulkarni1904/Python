#### Flattening and Raveling Arrays

Both `flatten()` and `ravel()` methods are used to convert a multi-dimensional array into a 1D array. However, there are key differences:

*   **`flatten()`**: Always returns a *copy* of the array. Changes to the flattened array will not affect the original array.
*   **`ravel()`**: Returns a *view* of the original array whenever possible. This means that changes to the raveled array might affect the original array. If the array is not contiguous in memory, `ravel()` might return a copy.

`ravel()` is generally more memory-efficient as it avoids creating a new copy if a view is possible.

```python
arr=np.array([[1,2,3],
                [4,5,6]])
arr.flatten()
```

```python
arr=np.array([[1,2,3],
                [4,5,6]])
arr.ravel()
```
