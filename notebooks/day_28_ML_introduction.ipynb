# Simple Linear Regression Example

```python
path = "/content/Salary_Data.csv"
```

### 1. Define Data Path

First, we specify the path to our dataset, which is `Salary_Data.csv`.

```python
import pandas as pd
```

### 2. Import Libraries

We import the necessary libraries, primarily `pandas` for data manipulation.

```python
# Load data_set
sal=pd.read_csv(path)
sal.head()
```

### 3. Load and Inspect Data

We load the CSV file into a pandas DataFrame and display the first few rows to understand its structure.

```python
# prepare data
X=sal[['YearsExperience']] #ML algorithms always take input in 2d

y=sal['Salary']
```

### 4. Prepare Data for Model

We separate the features (YearsExperience) and the target variable (Salary). For scikit-learn, features usually need to be in a 2D array.

```python
# anylysis
import matplotlib.pyplot as plt
plt.scatter(X,y,color='red',marker='*')
plt.title('Salary vs Experience')
plt.ylabel('Salary')
plt.xlabel('YearsExperience')
```

### 5. Visualize Raw Data

We create a scatter plot to visualize the relationship between 'YearsExperience' and 'Salary'. This helps in understanding if a linear relationship exists.

```python
# import algorithm --> scikit learn --> sklearn
# Paskel case to declear class where 1st letter is capital
from sklearn.linear_model import LinearRegression
```

### 6. Import Linear Regression Model

We import the `LinearRegression` class from scikit-learn's `linear_model` module.

```python
# as it is a class we need to creat a constructor --> obj
model=LinearRegression()
```

### 7. Instantiate the Model

We create an instance of the `LinearRegression` model.

```python
model
```

```python
# Training---> usinf fit() method
model.fit(X,y)
```

### 8. Train the Model

We train the linear regression model using our prepared `X` (features) and `y` (target) data. The `fit()` method calculates the optimal slope (m) and intercept (c) for the line equation (Y = mX + c).

```python
# Training---> means it is claculating the slope and intercept which is required for formation of line equation
# formation of the line equation
```

```python
# Testing
import warnings
warnings.filterwarnings('ignore')
model.predict([[0]])
```

### 9. Make Predictions

After training, we can use the model to predict salaries for new experience values. Here we test a few different values.

```python
model.predict([[7.5]])
```

```python
model.predict([[25]])
```

```python
sal.describe().T
```

### 10. Data Description

We can use `describe().T` to get a statistical summary of our dataset, including count, mean, standard deviation, min, max, and quartiles for each column.

# **Y=mx+c**

### 11. Extract Model Coefficients

We extract the slope (coefficient) `m` and the intercept `c` from our trained `LinearRegression` model. These are the parameters of the best-fit line.

```python
# m- slope
m = model.coef_
m
```

```python
# c - intercept
c = model.intercept_
c
```

```python
m*7.5+c
```

```python
m*11+c
```

### 12. Prepare Test Data for Plotting

We define a small set of experience years to use for plotting the regression line.

```python
test_data = [[0],[4.5],[6],[11]]
```

```python
y_pred = model.predict(test_data)
y_pred
```

### 13. Predict with Test Data

We use the trained model to predict salaries for our `test_data` points.

```python
# actual points
plt.scatter(X,y,color='red',marker='*')

# predicted line
plt.plot(test_data,y_pred,color='blue')
plt.title('Salary vs Experience')
plt.ylabel('Salary')
plt.xlabel('YearsExperience')
```

Finally, we plot the actual data points (red stars) and overlay the predicted regression line (blue) to visually assess the model's fit.
