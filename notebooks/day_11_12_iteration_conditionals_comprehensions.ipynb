## Iterative Flow Control

### Basic `for` Loop

The `for` loop is used to iterate over a sequence (like a list, tuple, dictionary, set, or string). Each item in the sequence is processed in order.

```python
for n in nms:
  if n.endswith('y'):
   print(n.upper())
```

### Conditional Iteration with String Methods

Here, we combine a `for` loop with an `if` condition to selectively perform actions. The `endswith('y')` method checks if a string ends with the specified character, and `upper()` converts the string to uppercase.

```python
#Fetch names in which i is present
for n in nms:
  if 'i' in n:
    print(n.title())
```

### Checking for Substring Presence

This example uses the `in` operator within an `if` condition to check if a character (`'i'`) is present in a string. The `title()` method capitalizes the first letter of each word.

```python
#fetch names whose length is <= 5
for n in nms:
  if len(n)<=5:
    print(n)
```

### Assignment: Fetch numbers divisible by 5 and 7

This section presents three ways to solve the problem of finding numbers divisible by both 5 and 7 from a list: using a traditional `for` loop, list comprehension with an `if` condition (for filtering), and list comprehension with an `if-else` condition (for transformation).

```python
nums=[23,10,70,45,90,100,35,8,16]

# Method 1: Using a traditional for loop (as provided by the user)
print('Using for loop:')
for n in nums:
  if (n%5==0)and(n%7==0):
    print(n)

# Method 2: Using list comprehension with an 'if' condition (filtering)
print('\nUsing list comprehension with if (filtering):')
divisible_numbers_lc = [n for n in nums if (n % 5 == 0) and (n % 7 == 0)]
print(divisible_numbers_lc)

# Method 3: Using list comprehension with an 'if-else' condition (transformation)
# This transforms the list, returning the number if it meets the condition, else 0.
print('\nUsing list comprehension with if-else (transformation):')
transformed_numbers_lc = [n if (n % 5 == 0) and (n % 7 == 0) else 0 for n in nums]
print(transformed_numbers_lc)
```

## List Comprehension

List comprehensions offer a concise way to create lists. They can be thought of as a single line `for` loop with optional `if` or `if-else` conditions. They are generally more readable and often more efficient than traditional `for` loops for list creation.

**Syntax:**
*   `[expression for item in iterable]`
*   `[expression for item in iterable if condition]`
*   `[expression_if_true if condition else expression_if_false for item in iterable]`

```python
#need a list of even numbers
[num for num in range (1,21)if num%2==1]
```

```python
#Ternary operator
#it can be operated on if , if..else
#syntax :
#express_1 if<condition>else express_2
```

## Ternary Operator

The ternary operator allows for a compact conditional expression. It's often used within list comprehensions or other assignments where a value needs to be chosen based on a simple condition.

**Syntax:** `value_if_true if condition else value_if_false`

```python
gender =['Male','Female','Female','Male']
#convert male to 0 & female to 1
[0 if val=='Male'else 1 for val in gender]
```

```python
salary = [24000,12500,45000,65000,11000,10000]
#Problem Statement: if sal>25000 then 10% bonus else 20%
[sal+(sal*0.1) if sal>25000 else sal+(sal*0.2) for sal in salary]
```

## Functions in Python

Functions are blocks of organized, reusable code that perform a single, related action. They help in breaking down a program into smaller, modular chunks, improving readability, reusability, and maintainability.

```python
#addition of 2 numbers
#syntax: deffunction_name():
            #pass
#Decleration
def addition():
  a=int(input("Enter first number:"))
  b=int(input("Enter second number:"))
  print('Total',a+b)
#only decleration doesnot provide the output
#to get the output we need to call the function_name()
addition()
```

### Function Declaration and Calling

-   **Declaration:** Defines the function, including its name, parameters (if any), and the body of the code it executes. Keyword `def` is used.
-   **Calling:** Executes the code inside the function. You call a function by its name followed by parentheses `()`.
