### Key Features of Python Lists

Python lists are versatile and come with several powerful features:

*   **Indexing and Slicing**: Elements can be accessed by their position (index), and sub-lists can be extracted using slicing.
*   **Accepts All Data Types**: Lists can store elements of any data type (integers, strings, other lists, objects, etc.).
*   **Homogeneous and Heterogeneous**: You can create lists with elements of the same type (homogeneous) or different types (heterogeneous).
*   **Preserves Sequence Order**: The order in which elements are added to a list is maintained.
*   **Allows Duplicate Records**: Lists can contain multiple instances of the same element.
*   **Mutable in Nature**: Lists are mutable, meaning their elements can be changed, added, or removed after the list has been created.

### Q. `append()` vs. `extend()`

*   **`append(object)`**: Adds its argument as a single element to the end of a list. The length of the list increases by one.
    *   Example: `list.append([1, 2])` results in `[..., [1, 2]]` (adds a list as one item).
*   **`extend(iterable)`**: Iterates over its argument and adds each element to the list, extending the list. The length of the list increases by the number of elements in the iterable.
    *   Example: `list.extend([1, 2])` results in `[..., 1, 2]` (adds elements 1 and 2 individually).

### Q. `pop()` vs. `remove()`

*   **`pop([index])`**: Removes and returns the item at the given `index`. If no index is specified, `pop()` removes and returns the last item in the list. It raises an `IndexError` if the list is empty or the index is out of range.
    *   Best for removing by position or the last element.
*   **`remove(value)`**: Removes the *first occurrence* of the specified `value` from the list. It raises a `ValueError` if the specified value is not found in the list.
    *   Best for removing a specific value when its position is unknown or not important.

### List Copying: Shallow Copy vs. Reference

*   **Shallow Copy (`.copy()`):** Creates a *new* list that contains the same elements as the original. Changes to the new list (like adding or removing elements) do not affect the original list, and vice-versa. However, if the elements themselves are mutable objects (e.g., lists within lists), changes to those nested mutable objects will be reflected in both lists, as they still refer to the same nested objects.
    *   Example: `x = t.copy()` creates a new list `x` independent of `t` for top-level elements.

*   **Reference (Direct Assignment):** Assigning one list to another variable (e.g., `y = x`) does *not* create a new list. Instead, both variables (`x` and `y`) point to the *exact same list object* in memory. Any modification made through `x` will be visible through `y`, and vice-versa.
    *   Example: `y = x` means `x` and `y` are just two names for the same list.

### Assignment: Can a Tuple be added to a List using `extend()`?

No, `extend()` cannot add an entire tuple as a single element. The `extend()` method expects an iterable and adds each *individual element* from that iterable to the list. If you want to add a tuple as a single element, you should use `append()`.

Let's confirm this with an example:

````python
my_list = [1, 2, 3]
my_tuple = (4, 5, 6)

print(f"Original list: {my_list}")

# Using extend() with a tuple
list_extended = list(my_list) # Create a copy to show extend
list_extended.extend(my_tuple)
print(f"List after extend() with a tuple: {list_extended} (elements of tuple are added individually)")

# Using append() with a tuple
list_appended = list(my_list) # Create a copy to show append
list_appended.append(my_tuple)
print(f"List after append() with a tuple: {list_appended} (the entire tuple is added as one element)")
````

## 1. **`os`** — Operating System Interface

**Theory:** Provides functions to interact with the operating system — file paths, directories, and environment variables. It allows you to navigate folders, create/rename/delete directories, and build cross-platform paths.

```python
# =========================
# 1. OS MODULE
# =========================
import os

print("Current Dir:", os.getcwd())
print("Files:", os.listdir("."))
print("Exists:", os.path.exists("test.txt"))
print("Path:", os.path.join("folder", "file.txt"))
print("Home directory:", os.environ.get('HOME'))

# Create a directory safely (won't crash if it already exists — important in Colab re-runs)
if not os.path.exists("test_folder"):
    os.mkdir("test_folder")
print("After mkdir:", os.listdir("."))
```

## 2. **`io`** — Input/Output Streams

**Theory:** Provides tools for working with streams (files, in-memory buffers). `StringIO` creates a text file *in memory* and `BytesIO` does the same for binary data — useful when you want file-like behavior without creating an actual file.

```python
# =========================
# 2. IO MODULE
# =========================
import io

s = io.StringIO()
s.write("Hello Python")
print("StringIO:", s.getvalue())
s.close()

b = io.BytesIO()
b.write(b"Binary data here")
print("BytesIO:", b.getvalue())
b.close()
```

## 3. **`sys`** — System-Specific Parameters

## 4. **`array`** — Efficient Arrays of Numeric Values

**Theory:** Stores basic numeric types (like C-style arrays) more compactly than lists. Unlike lists, all elements must be the **same type**, making them more memory-efficient for large numeric data.

**Type codes:** `'i'` (int), `'f'` (float), `'d'` (double), `'b'` (signed char), etc.

```python
# =========================
# 4. ARRAY MODULE
# =========================
import array

arr = array.array('i', [10, 20, 30])
arr.append(40)
print("Array:", arr)
print("First Element:", arr[0])
print("Type code 'i' = signed int")
```

## 5. **`csv`** — CSV File Reading/Writing

**Theory:** Reads and writes Comma-Separated Values files — the standard format for spreadsheet data exchange. `csv.reader` gives rows as lists, while `csv.DictReader` gives rows as dictionaries (column-based access).

```python
# =========================
# 5. CSV MODULE
# =========================
import csv

# Writing
with open("sample.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Name", "Age"])
    writer.writerow(["Akshay", 22])

# Reading
with open("sample.csv", "r") as f:
    for row in csv.reader(f):
        print(row)

# DictReader (access by column name)
with open("sample.csv", "r") as f:
    for row in csv.DictReader(f):
        print(row["Name"], row["Age"])
```

## 6. **`openpyxl`** — Excel File Handling

**Theory:** Reads and writes `.xlsx` Excel files (third-party module: `pip install openpyxl`). Lets you create workbooks, write to cells, and read sheets programmatically.

```python
# =========================
# 6. OPENPYXL MODULE
# =========================
from openpyxl import Workbook, load_workbook

# Create & write
wb = Workbook()
ws = wb.active
ws["A1"] = "Name"
ws["B1"] = "Marks"
ws.append(["Akshay", 90])
wb.save("data.xlsx")   # Saved file appears in Colab's file panel (left sidebar )

# Load & read back
wb = load_workbook("data.xlsx")
ws = wb.active
print("Excel Cell A1:", ws["A1"].value)
for row in ws.iter_rows(values_only=True):
    print(row)
```

## 7. **`collections`** — Specialized Container Types

**Theory:** Provides alternatives to built-in containers with extra features:
- **Counter** → counts occurrences of elements
- **defaultdict** → dict that never raises `KeyError` (auto-default values)
- **deque** → fast appends/pops from **both** ends

```python
# =========================
# 7. COLLECTIONS MODULE
# =========================
from collections import Counter, defaultdict, deque

print("Counter:", Counter("banana"))

d = defaultdict(int)
d["A"] += 1          # No KeyError — defaults to 0
print("DefaultDict:", dict(d))

dq = deque([1, 2, 3])
dq.appendleft(0)
print("Deque:", dq)
```

## 8. **`itertools`** — Iterator Building Blocks

**Theory:** Creates efficient iterators for looping, combining, and filtering data. `combinations` gives unique selections (order doesn't matter), while `permutations` gives all orderings (order matters).

```python
# =========================
# 8. ITERTOOLS MODULE
# =========================
from itertools import combinations, permutations

print("Combinations:", list(combinations([1, 2, 3], 2)))
print("Permutations:", list(permutations([1, 2, 3], 2)))
```

## 9. **`json`** — JSON Encoding/Decoding

**Theory:** Parses JSON strings and converts Python objects to/from JSON — essential for APIs and config files.
- `dumps()` → Python → JSON string
- `loads()` → JSON string → Python

```python
# =========================
# 9. JSON MODULE
# =========================
import json

data = {"name": "Akshay", "age": 22}

json_str = json.dumps(data)
print("JSON String:", json_str)

python_obj = json.loads(json_str)
print("Python Object:", python_obj)
```
