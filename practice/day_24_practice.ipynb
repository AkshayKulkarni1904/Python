```python
path='/content/CO2 Emissions_Canada.csv'
```

```python
import pandas as pd
c=pd.read_csv(path)
```

```python
c['Make'].unique()
```

```python
c.columns
```

```python
#Display all unique fuel types
c["Fuel Type"].unique()
```

```python
#Find the average CO2 emission of all vehicles
print(c['CO2 Emissions(g/km)'].mean())
```

```python
#For rounded output:
print(round(c['CO2 Emissions(g/km)'].mean(),2))
```

```python
#Find the vehicle with the highest CO2 emission
c.loc[c['CO2 Emissions(g/km)'].max()]
```

```python
#Find the vehicle with the lowest CO2 emission
c.loc[c['CO2 Emissions(g/km)'].min()]
```

```python
#Find all vehicles having engine size greater than 4 litres
c.query("`Engine Size(L)`>4")
```

```python
#find the total number:
len(c[c["Engine Size(L)"] > 4])
```

```python
#Count the number of vehicles for each manufacturer
c['Make'].value_counts()
```

```python
#Find the average CO2 emission for each fuel type
c.groupby("Fuel Type")["CO2 Emissions(g/km)"].mean()
```

```python
#Find the top 5 manufacturers with the highest average CO2 emission
c.groupby('Make')['CO2 Emissions(g/km)'].mean().sort_values(ascending=False).head()
```

```python
#Find all BMW vehicles with CO2 emissions greater than 300 g/km
c[(c['Make']=='BMW')&(c['CO2 Emissions(g/km)']>300)]
```

```python
#Sort the dataset based on CO2 emissions from highest to lowest
c.sort_values(by='CO2 Emissions(g/km)',ascending=False)
```

```python
#top 10
c.sort_values(by='CO2 Emissions(g/km)',ascending=False).head(10)
```

```python
#Find the correlation between engine size and CO2 emissions
c['Engine Size(L)'].corr(c['CO2 Emissions(g/km)'])
```
