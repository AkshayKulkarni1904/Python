```python
path="/content/CO2 Emissions_Canada.csv"
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
c=pd.read_csv(path)
c
```

```python
c.info
```

```python
c.shape
```

```python
c.columns
```

```python
#matplotlib operations
plt.plot(c['Engine Size(L)'][:100],c['CO2 Emissions(g/km)'][:100],'m^')
plt.xlabel('Engine Size(L)')
plt.ylabel('CO2 Emissions(g/km)')
plt.grid()
plt.title('Engine Size vs CO2 Emissions')
plt.show()
```

```python
#scatter plot
plt.scatter(c["Engine Size(L)"],
            c["CO2 Emissions(g/km)"])

plt.xlabel("Engine Size (L)")
plt.ylabel("CO2 Emissions (g/km)")
plt.title("Engine Size vs CO2 Emissions")

plt.show()
```

```python
#bar plt
fuel_co2 = c.groupby("Fuel Type")["CO2 Emissions(g/km)"].mean()

plt.bar(fuel_co2.index, fuel_co2.values)

plt.xlabel("Fuel Type")
plt.ylabel("Average CO2 Emissions")
plt.title("Average CO2 Emissions by Fuel Type")

plt.show()
```

```python
fuel_co2 = c.groupby("Fuel Type")["CO2 Emissions(g/km)"].mean()

plt.barh(fuel_co2.index, fuel_co2.values)

plt.xlabel("Average CO2 Emissions")
plt.ylabel("Fuel Type")
plt.title("CO2 Emissions by Fuel Type")

plt.show()
```

```python
#histogram
plt.hist(c["CO2 Emissions(g/km)"], bins=20)

plt.xlabel("CO2 Emissions")
plt.ylabel("Frequency")
plt.title("Distribution of CO2 Emissions")

plt.show()
```

```python
#pie chart
fuel_count = c["Fuel Type"].value_counts()

plt.pie(fuel_count,
        labels=fuel_count.index,
        autopct="%1.1f%%")

plt.title("Fuel Type Distribution")

plt.show()
```

```python
#box plot - CO2
plt.boxplot(c["CO2 Emissions(g/km)"])

plt.ylabel("CO2 Emissions")
plt.title("CO2 Emissions Box Plot")

plt.show()
```

```python
# Multiple Line Plot — City and Highway Fuel Consumption
plt.plot(c["Fuel Consumption City (L/100 km)"][:100],
         label="City")

plt.plot(c["Fuel Consumption Hwy (L/100 km)"][:100],
         label="Highway")

plt.xlabel("Vehicle Index")
plt.ylabel("Fuel Consumption")
plt.title("City vs Highway Fuel Consumption")

plt.legend()

plt.show()
```

```python
#change fig size
plt.figure(figsize=(10, 5))

plt.scatter(c["Engine Size(L)"],
            c["CO2 Emissions(g/km)"])

plt.xlabel("Engine Size")
plt.ylabel("CO2 Emissions")
plt.title("Engine Size vs CO2 Emissions")

plt.show()
```

```python
plt.style.use("Solarize_Light2")

plt.scatter(
    c["Engine Size(L)"],
    c["CO2 Emissions(g/km)"]
)

plt.title("Engine Size vs CO2 Emissions")

plt.show()
```

```python
#Top 10 Vehicle Manufacturers Bar Chart
top_make = c["Make"].value_counts().head(10)

plt.figure(figsize=(10, 5))

plt.bar(top_make.index, top_make.values)

plt.xlabel("Vehicle Make")
plt.ylabel("Number of Vehicles")
plt.title("Top 10 Vehicle Manufacturers")

plt.xticks(rotation=45)

plt.show()
```

#**Seaborn Operations**

```python
#Scatter Plot
sns.scatterplot(x="Engine Size(L)",y="CO2 Emissions(g/km)",data=c, color="green",
    marker="x")

plt.title("Engine Size vs CO2 Emissions")

plt.show()
```

```python
#Line Plot
sns.lineplot(x="Engine Size(L)",y="CO2 Emissions(g/km)",data=c, color="blue",
    alpha=0.3)

plt.title("Engine Size vs CO2 Emissions")

plt.show()
```

```python
#Bar Plot
sns.barplot(x="Fuel Type",y="CO2 Emissions(g/km)",data=c)

plt.title("Fuel Type vs CO2 Emissions")

plt.show()
```

```python
#Count Plot
sns.countplot(x="Fuel Type",data=c)
plt.title("Fuel Type Distribution")

plt.show()
```

```python
#Histogram
sns.histplot(c["CO2 Emissions(g/km)"],bins=20)

plt.title("Distribution of CO2 Emissions")

plt.show()
```

```python
#Histogram with KDE
sns.histplot(c["CO2 Emissions(g/km)"],kde=True)

plt.title("Distribution of CO2 Emissions")

plt.show()
```

```python
#KDE Plot
sns.kdeplot(c["CO2 Emissions(g/km)"])

plt.title("Kernel Density Estimation of CO2 Emissions")

plt.show()
```

```python
#Box Plot
sns.boxplot(x="Fuel Type",y="CO2 Emissions(g/km)",data=c)

plt.title("Box Plot of CO2 Emissions by Fuel Type")

plt.show()
```

```python
#Violin Plot
sns.violinplot(x="Fuel Type",y="CO2 Emissions(g/km)",data=c)

plt.title("Violin Plot of CO2 Emissions by Fuel Type")

plt.show()
```

```python
#Regression Plot
sns.regplot(x="Engine Size(L)",y="CO2 Emissions(g/km)",data=c)

plt.title("Scatter Plot with Regression Line")

plt.show()
```

```python
# Correlation Heatmap
numeric_c = c.select_dtypes(include=["number"])

correlation = numeric_c.corr()

plt.figure(figsize=(10, 6))

sns.heatmap(
    correlation,
    annot=True
)

plt.title("Correlation Heatmap")
plt.xticks(rotation=45)

plt.show()
```

```python
# Pair Plot
plt.figure(figsize=(10, 15))

sns.pairplot(c)

plt.show()
columns = [
    "Engine Size(L)",
    "Cylinders",
    "Fuel Consumption Comb (L/100 km)",
    "CO2 Emissions(g/km)"
]

sns.pairplot(c[columns])

plt.show()
```

```python
#Strip Plot
sns.stripplot(
    x="Fuel Type",
    y="CO2 Emissions(g/km)",
    data=c
)

plt.title("CO2 Emissions by Fuel Type")

plt.show()
```

```python
sns.set_theme(style="darkgrid")


sns.scatterplot(
    x="Engine Size(L)",
    y="CO2 Emissions(g/km)",
    data=c
)

plt.show()
```

```python
sns.barplot(
    x="Fuel Type",
    y="CO2 Emissions(g/km)",
    data=c,
    palette="magma",
    hue="Fuel Type",
    legend=False
)

plt.show()
```

```python
#Swarm Plot
sample_df = c.sample(500)

sns.swarmplot(
    x="Fuel Type",
    y="CO2 Emissions(g/km)",
    data=sample_df
)

plt.title("CO2 Emissions by Fuel Type")

plt.show()
```

```python
#Joint Plot
sns.jointplot(x="Engine Size(L)",y="CO2 Emissions(g/km)",data=c)

plt.show()
```

```python
#Boxen Plot
sns.boxenplot(x="Fuel Type",y="CO2 Emissions(g/km)",data=c)

plt.title("Boxen Plot of CO2 Emissions by Fuel Type")

plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.scatter(
    c["Engine Size(L)"],
    c["CO2 Emissions(g/km)"],
    color="purple",
    marker="D",
    s=50,
    alpha=0.5
)

plt.xlabel("Engine Size")
plt.ylabel("CO2 Emissions")
plt.title("Engine Size vs CO2 Emissions")

plt.grid()

plt.show()
```
