## 3. Grocery Bill Examples

This section provides various approaches to creating a simple grocery bill, demonstrating different ways to take multiple inputs from the user and calculate a total.

### Single Item Grocery Bill

This example calculates the bill for a single item by asking for its name, price, and quantity. It's a basic illustration of taking individual inputs and multiplying them for a subtotal.

```python
#Grocery bill
#Single item
item_name=input('Enter item name:')
item_price=float(input('Enter item price:'))
item_quantity=float(input('Enter item quantity:'))

print('Final Bill is Rs.',item_price*item_quantity)
```

### Multiple Items Grocery Bill (Individual Inputs)

This cell demonstrates how to handle multiple items by taking their names and prices separately. It uses `split()` to separate space-delimited inputs and `map(float, ...)` to convert price strings to floating-point numbers before summing them up.

```python
#Multiple items
item1, item2, item3 = input("Enter 3 items separated by spaces: ").split()

price1, price2, price3 = map(float, input("Enter 3 prices separated by spaces: ").split())

total = price1 + price2 + price3

print("\nItems:", '\n',item1, '\n',item2, '\n',item3)
print("Total Bill =", total)
```

### Multiple Items Grocery Bill (Using `map()` and `sum()`)

This is a more efficient method for handling multiple items. It uses `split(',')` for item names and `map(float, ...)` combined with `list()` to process prices. The `sum()` function is then used to easily calculate the total bill from the list of prices.

```python
# Other method is to use map function
# map() is a built-in Python function that applies a function to every element in a collection (like a list, tuple, or input values).
# syntax: map(function, iterable)
items = input("Enter item names separated by commas: ").split(",")
prices = list(map(float, input("Enter prices separated by spaces: ").split()))

total = sum(prices)

print("\n----- GROCERY BILL -----")
print("Items:", items)
print("Prices:", prices)
print("Total Bill Amount =", total)
```

## 4. Advanced Input Handling with `eval()`

This section introduces the `eval()` function as a way to take complex data structures (like lists, sets, tuples, or dictionaries) directly as input from the user.

**Note**: While `eval()` is powerful, it should be used with caution in production environments as it can execute arbitrary code if not handled carefully.

### Demonstrating `eval(input())`

This cell uses `eval(input())` to allow the user to enter various data types directly. It then prints the entered data and its inferred Python type, showcasing the flexibility of `eval()`.

```python
data = eval(input("Enter you data:"))

print(data)
print(type(data))
```
