### Python Dynamic Typing

Unlike statically typed languages where you declare the variable type explicitly (e.g., `int a = 15`), Python uses dynamic typing. This means the type of a variable is determined at runtime based on the value assigned to it.

```python
# Example of dynamic typing
a = 20
print(f"The value of 'a' is {a} and its type is {type(a)}")

a = "Hello Python"
print(f"Now, the value of 'a' is '{a}' and its type is {type(a)}")
```

## Mutable vs. Immutable Data Types

In Python, understanding whether a data type is *mutable* or *immutable* is crucial. Mutable objects can be changed after they are created, while immutable objects cannot. This distinction has significant implications for how data is handled in your programs.

### Lists

Lists are ordered, mutable collections of items. They are defined using square brackets `[]` and can contain elements of different data types.

```python
# Examples of Lists
q = [1, 2, 3] # Homogeneous list of integers
r = ['a', 'b', 'c'] # Homogeneous list of strings
p = ['X', 10, 20.5, 'Y'] # Heterogeneous list

print(f"Type of q: {type(q)}, r: {type(r)}, p: {type(p)}")
```

### Tuples

Tuples are ordered, immutable collections of items. They are defined using parentheses `()` and, like lists, can hold elements of various data types.

```python
# Examples of Tuples
t1 = (1, 2, 3) # Homogeneous tuple of integers
t2 = ('a', 'b', 'c') # Homogeneous tuple of strings
t3 = ('X', 10, 20, 30, 'Y') # Heterogeneous tuple

print(f"Type of t1: {type(t1)}, t2: {type(t2)}, t3: {type(t3)}")
```

## Assignment: List vs. Tuple

To solidify our understanding, let's explore the key differences between Lists and Tuples, especially concerning their mutability.

### List vs. Tuple: Detailed Explanation

#### List (`[]`)
*   **Mutable**: Lists can be changed after they are created. You can add, remove, or modify elements.
*   **Dynamic Size**: Their size can change.
*   **Slower**: Due to their mutability, operations on lists can be slightly slower than tuples for certain operations.
*   **Use Cases**: When you need a collection of items that might change over time, such as a shopping cart, a record of events, or data that needs frequent updates.

#### Tuple (`()`)
*   **Immutable**: Tuples cannot be changed after they are created. Once defined, their elements and size are fixed.
*   **Fixed Size**: Their size is constant.
*   **Faster**: Due to their immutability, operations on tuples are generally faster than lists.
*   **Use Cases**: When you need a collection of items that should not change, such as coordinates (x, y), RGB color values, or database records. Also used for returning multiple values from a function or as keys in dictionaries (if all elements are immutable).

```python
# --- List Examples (Mutable) ---
print('--- List Examples (Mutable) ---')
my_list = [1, 2, 3]
print(f"Original List: {my_list}")

# Modifying a list: Adding an element
my_list.append(4)
print(f"List after append: {my_list}")

# Modifying a list: Changing an element
my_list[0] = 10
print(f"List after modifying element: {my_list}")

# --- Tuple Examples (Immutable) ---
print('\n--- Tuple Examples (Immutable) ---')
my_tuple = (1, 2, 3)
print(f"Original Tuple: {my_tuple}")

# Attempting to modify a tuple will raise an error
try:
    my_tuple[0] = 10
except TypeError as e:
    print(f"Error when trying to modify tuple element: {e}")

# Attempting to append to a tuple will raise an error
try:
    my_tuple.append(4)
except AttributeError as e:
    print(f"Error when trying to append to tuple: {e}")
```

## Memory Management: Shallow vs. Deep Copy

When you copy complex objects (like lists containing other lists), Python offers two main ways to duplicate them: shallow copy and deep copy. The difference lies in how they handle nested objects.

### Shallow Copy

A shallow copy creates a new compound object and then inserts references into it to the objects found in the original. It does not create copies of nested objects.

*   If the original object contains nested objects (e.g., a list within a list), the shallow copy will refer to the *same* nested objects as the original.
*   Changes to mutable nested objects in either the original or the shallow copy will be reflected in both.
*   Achieved using `list.copy()`, `dict.copy()`, slicing (`[:]`), or `copy.copy()` from the `copy` module.

### Deep Copy

A deep copy creates a new compound object and then, recursively, inserts copies of the objects found in the original into the new object. It creates independent copies of all nested objects.

*   A deep copy creates a completely independent clone, including all nested mutable objects.
*   Changes to mutable nested objects in the deep copy will *not* affect the original, and vice-versa.
*   Achieved using `copy.deepcopy()` from the `copy` module.

```python
import copy

# Original nested list
original_list = [[1, 2], [3, 4]]
print(f"Original List: {original_list}, ID: {id(original_list)}")
print(f"ID of nested list in original: {id(original_list[0])}")

# --- Shallow Copy ---
shallow_copied_list = copy.copy(original_list)
print(f"\nShallow Copied List: {shallow_copied_list}, ID: {id(shallow_copied_list)}")
print(f"ID of nested list in shallow copy: {id(shallow_copied_list[0])}")

# Modify an element in the nested list of the shallow copy
shallow_copied_list[0][0] = 99

print(f"Original List after shallow copy modification: {original_list}")
print(f"Shallow Copied List after modification: {shallow_copied_list}")

# --- Deep Copy ---
original_list_deep = [[5, 6], [7, 8]]
print(f"\nOriginal List for deep copy: {original_list_deep}, ID: {id(original_list_deep)}")
print(f"ID of nested list in original (deep): {id(original_list_deep[0])}")

deep_copied_list = copy.deepcopy(original_list_deep)
print(f"Deep Copied List: {deep_copied_list}, ID: {id(deep_copied_list)}")
print(f"ID of nested list in deep copy: {id(deep_copied_list[0])}")

# Modify an element in the nested list of the deep copy
deep_copied_list[0][0] = 999

print(f"Original List after deep copy modification: {original_list_deep}")
print(f"Deep Copied List after modification: {deep_copied_list}")
```

### Explanation of Shallow Copy Output

As you can see from the example above:

*   The `original_list` and `shallow_copied_list` are different objects (they have different IDs).
*   However, their nested lists (`[1, 2]` in the `original_list` and `[3, 4]` in the `shallow_copied_list`) are the *same* objects in memory (they have the same ID).
*   When `shallow_copied_list[0][0]` was changed to `99`, it also affected `original_list[0][0]` because both lists were pointing to the *same* nested list object.
*   This clearly demonstrates that a shallow copy shares references to nested mutable objects.

### Explanation of Deep Copy Output

From the deep copy example, we observe:

*   The `original_list_deep` and `deep_copied_list` are distinct objects.
*   Crucially, their nested lists (`[5, 6]` and `[7, 8]`) are also different objects in memory (they have different IDs).
*   When `deep_copied_list[0][0]` was changed to `999`, it did not affect `original_list_deep[0][0]` because `deepcopy` created an entirely separate copy of the nested list as well.

In summary, choose `copy.copy()` for shallow copies when you're okay with shared nested mutable objects, and `copy.deepcopy()` for deep copies when you need a completely independent clone, including all nested mutable objects.

## Python Control Flow Statements

Control flow statements are essential for directing the order in which instructions are executed in a program. They allow you to make decisions, repeat actions, and handle specific conditions.

### The `break` Statement

The `break` statement is used to terminate the loop entirely. When `break` is encountered, the loop is immediately exited, and program execution continues from the statement immediately after the loop.

```python
print('--- Example of `break` ---')
for i in range(5):
  if i == 3:
    print(f"Breaking loop at i = {i}")
    break
  print(f"Current value of i: {i}")
print("Loop finished.")
```

### The `continue` Statement

The `continue` statement is used to skip the rest of the current iteration of the loop and move to the next iteration. The loop does not terminate; it just bypasses the current step.

```python
print('--- Example of `continue` ---')
for i in range(7):
  if i == 3 or i == 5:
    print(f"Skipping i = {i} with continue")
    continue
  print(f"Current value of i: {i}")
print("Loop finished.")
```

### The `pass` Statement

The `pass` statement is a null operation; nothing happens when it executes. It is used as a placeholder where a statement is syntactically required but you do not want any command or code to execute. This can be useful for functions, classes, or loops that you haven't implemented yet, preventing `SyntaxError`.

```python
print('--- Example of `pass` in a conditional statement ---')
n = 10

# using pass inside an if statement as a placeholder
if n > 10:
  pass  # Do nothing if n is greater than 10
  print("This line will only execute if the 'if' block is entered and 'pass' is the only statement before it.")
else:
  print(f"n is {n}, which is less than or equal to 10.")

print('\n--- Example of `pass` in a function definition ---')
def future_function():
  pass # Placeholder for future implementation

future_function()
print("future_function was called and did nothing as expected.")
```

## Python Functions

Functions are blocks of organized, reusable code that perform a single, related action. They help break down our program into smaller, modular chunks, making the code more readable, maintainable, and reusable.

### Defining and Calling a Function

Functions are defined using the `def` keyword, followed by the function name, parentheses `()`, and a colon `:`. The function's code block is indented.

```python
def greet():
  print("Hello from the greet function!")

# Function call
greet()
print("This line is outside the function.")
```

### Function Arguments

Functions can accept arguments, which are values passed into the function to perform its task. These arguments act as variables within the function's scope.

```python
def greet_person(name):
  print(f"Hello, {name}!")

greet_person("Akshay")
greet_person("Alice")
```

### Function to Add Two Numbers (with arguments)

```python
def add_numbers(num1, num2):
  sum_result = num1 + num2
  print(f"The sum of {num1} and {num2} is: {sum_result}")

add_numbers(15, 4)
add_numbers(100, 250)
```

### The `return` Statement

The `return` statement is used to send a value back from a function to the calling part of the program. It also signifies the end of the function's execution.

```python
def find_square(num):
  result = num * num
  return result

# Call the function and store the returned value
square_of_54 = find_square(54)
print(f"The square of 54 is: {square_of_54}")

square_of_7 = find_square(7)
print(f"The square of 7 is: {square_of_7}")
```
