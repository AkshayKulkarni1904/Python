## Dictionaries

Dictionaries are unordered collections of data values, used to store data values like a map, which, unlike other data types that hold only a single value as an element, holds `key:value` pairs. Each `key:value` pair in a dictionary is separated by a comma. They are enclosed within curly braces `{}`. Keys must be unique and immutable (e.g., strings, numbers, tuples), while values can be of any data type and can be duplicated.

### Creating and Accessing Dictionaries

Here's how to define a dictionary and access its elements using keys.

```python
# Dictionary creation: {key: value}
d1 = {1: 100, 2: 200, 3: 300}

# Display the dictionary
print("Dictionary d1:", d1)

# Accessing values using keys
print("Value associated with key 1:", d1[1])
print("Value associated with key 3:", d1[3])
```

### Checking Dictionary Type

You can verify the type of the created object using the `type()` function.

```python
# Check the type of d1
print("Type of d1:", type(d1))
```

### Modifying and Adding Elements

Dictionaries are mutable, meaning you can add new key-value pairs or change existing values after creation.

```python
# Adding a new key-value pair
d1[4] = 400
print("Dictionary after adding a new element:", d1)

# Modifying an existing value
d1[2] = 250
print("Dictionary after modifying an element:", d1)
```

## Strings

Strings are fundamental data types in Python used to represent text. They are sequences of characters.

```python
# Example of a string
s = 'python'
print(s)
```

### Immutability of Strings

An important characteristic of strings in Python is that they are **immutable**. This means that once a string is created, its individual characters cannot be changed or updated. Any operation that seems to 'modify' a string actually creates a new string.

```python
my_string = "hello"
# Attempting to change a character will result in an error
# my_string[0] = 'H' # Uncommenting this line would raise a TypeError

# Operations like concatenation create new strings
new_string = my_string + " world"
print("Original string:", my_string)
print("New string after concatenation:", new_string)
```

### String as a Sequence (Array-like Structure)

Although strings are immutable, you can think of them as an array of characters, where each character has a specific position or index. This allows for accessing individual characters or parts of the string.

```python
# Each character in the string 'python' has an index:
# p ---> 0
# y ---> 1
# t ---> 2
# h ---> 3
# o ---> 4
# n ---> 5
```

### Indexing in Strings

Python supports two types of indexing to access characters in a string:

1.  **Positive Indexing**: Starts from `0` for the first character and goes up to `n-1` for the last character (from left to right).
2.  **Negative Indexing**: Starts from `-1` for the last character and goes down to `-n` for the first character (from right to left).

Indexing allows you to access a single character at a time.

```python
s = 'PYTHON'

# Positive indexing
print("Character at index 0 (positive):", s[0]) # Output: P
print("Character at index 3 (positive):", s[3]) # Output: H

# Negative indexing
print("Character at index -1 (negative):", s[-1]) # Output: N
print("Character at index -5 (negative):", s[-5]) # Output: Y
```

### Slicing in Strings

Slicing allows you to extract a substring (a part of the string) from a larger string. The syntax is `s[start:stop:step]`.

*   `start`: The starting index (inclusive). If omitted, defaults to 0.
*   `stop`: The ending index (exclusive). If omitted, defaults to the end of the string.
*   `step`: The increment between indices (defaults to 1). If negative, slices backward.

```python
s = 'PYTHON'

# Basic slicing: s[start:stop]
print("s[2:-1] (from index 2 up to, but not including, the last char):", s[2:-1]) # Output: THO
print("s[2:5] (from index 2 up to, but not including, index 5):", s[2:5])   # Output: THO

# Access 'PY'
print("s[:2] (from beginning up to, but not including, index 2):", s[:2]) # Output: PY

# Access 'YTH'
print("s[1:-2] (from index 1 up to, but not including, the second to last char):", s[1:-2]) # Output: YTH

# Access 'HO'
print("s[3:-1] (from index 3 up to, but not including, the last char):", s[3:-1]) # Output: HO
```

### Slicing with Step

The `step` parameter allows you to skip characters while slicing.

```python
s = 'PYTHON'

# s[::2] (start from beginning, go to end, step by 2)
print("s[::2] (every second character):", s[::2]) # Output: PTO

address = "Hanuman Nagar Kolhapur Pin-411007"
# Example: Extracting 'Nagar Kolh' from address using negative slicing
print("address[-25:-15]:", address[-25:-15]) # Output: Nagar Kolh
```

### Reversing a String

A common use of slicing with a negative step is to reverse a string.

```python
n = 'Akshay Kulkarni'

# Reverse string using slicing with step -1
print("Original string:", n)
print("Reversed string:", n[::-1]) # Output: inrakluK yahskA
```

## String Methods

Python strings come with a variety of built-in methods that allow you to perform common operations like changing case, stripping whitespace, finding substrings, and more. These methods return new strings, as strings are immutable.

```python
s = 'PYTHON'
```

### Case Conversion Methods

*   `lower()`: Converts all uppercase characters in a string to lowercase.
*   `capitalize()`: Converts the first character of the string to a capital letter and all other capital letters to lowercase.
*   `title()`: Converts the first character of each word to uppercase and remaining to lowercase.
*   `swapcase()`: Converts all uppercase characters to lowercase and all lowercase characters to uppercase.

```python
# Convert string to lower case
print("s.lower():", s.lower()) # Output: python

# Capitalize the first letter
print("s.capitalize():", s.capitalize()) # Output: Python

# Title case (first letter of each word capitalized)
print("'akshay kulkarni'.title():", 'akshay kulkarni'.title()) # Output: Akshay Kulkarni

h = 'gooD MoRninG'
print("Original string 'h':", h)
# Swap case
print("h.swapcase():", h.swapcase()) # Output: GOOd mOrNINg
```

### Stripping Whitespace

*   `strip()`: Removes leading and trailing whitespace (or specified characters) from a string.
*   `lstrip()`: Removes leading whitespace (or specified characters).
*   `rstrip()`: Removes trailing whitespace (or specified characters).

```python
# To remove spaces present in prefix and suffix using strip() method
print("'             python  '.strip():", '             python  '.strip()) # Output: python

# Remove specific characters from prefix and suffix
print("'*******Java******'.strip('*'):", '*******Java******'.strip('*')) # Output: Java
```

### Assignment: Removing Internal Spaces

**Task:** Remove spaces present within the string `'   J  A  V  A  '`. (Hint: `strip()` only removes leading/trailing spaces. For internal spaces, you might need different techniques, e.g., `replace()` or `split()` and `join()`, or regular expressions if complex.)

```python
text_with_internal_spaces = '   J  A  V  A  '
print("Original string:", f"'{text_with_internal_spaces}'")

# Method 1: Using replace()
# This removes all spaces, including internal ones.
no_internal_spaces_replace = text_with_internal_spaces.replace(' ', '')
print("Using replace():", f"'{no_internal_spaces_replace}'")

# Method 2: Using split() and join()
# First strip leading/trailing, then split by space and join without space
no_internal_spaces_join = ''.join(text_with_internal_spaces.strip().split())
print("Using split() and join():", f"'{no_internal_spaces_join}'")

# Note: If the goal is *only* to remove extra spaces between words while keeping single spaces,
# then `split()` followed by `join(' ')` would be used after stripping leading/trailing.
# E.g., '   Hello   World   '.split() -> ['Hello', 'World'] -> ' '.join(['Hello', 'World']) -> 'Hello World'
```

### Assignment: `casefold()` vs `lower()`

**Task:** Check the `casefold()` method and explain the difference between `casefold()` and `lower()`.

```python
str1 = "Straße"
str2 = "ẞTRASSE"

# Using lower()
print("'{str1}'.lower(): {str1.lower()}") # 'straße'
print("'{str2}'.lower(): {str2.lower()}") # 'ẞtrasse'

# Using casefold()
print("'{str1}'.casefold(): {str1.casefold()}") # 'strasse'
print("'{str2}'.casefold(): {str2.casefold()}") # 'strasse'

# Explanation:
# Both `lower()` and `casefold()` convert strings to lowercase.
# The key difference is that `casefold()` is a more aggressive method for converting strings to caseless matching.
# It handles characters that are not handled by `lower()`, especially in non-English languages.
# For example, the German Eszett ('ß') converts to 'ss' with `casefold()`, but remains 'ß' with `lower()`.
# Therefore, `casefold()` is generally preferred for case-insensitive comparisons, especially when dealing with international text.
```

### Checking String Prefixes and Suffixes

*   `startswith()`: Checks if the string starts with the specified prefix.
*   `endswith()`: Checks if the string ends with the specified suffix.

Both methods return a boolean (`True` or `False`).

```python
code = 'cbr1:itsarandomgibrishtoseemethod'

# Check if the string starts with 'cbr1'
print("code.startswith('cbr1'):", code.startswith('cbr1')) # Output: True

# Check if the string ends with 'ABCD'
print("code.endswith('ABCD'):", code.endswith('ABCD')) # Output: False
```

## Functions

Functions are blocks of reusable code that perform a specific task. They help organize code, make it more readable, and promote code reuse.

### Positional Arguments

Positional arguments are arguments that are passed into a function in the order they are defined in the function signature. The position of the argument matters, and the values are assigned to parameters based on their order.

Consider the `display_info` function. If called with `display_info('Akshay', 'Kulkarni')`, then `'Akshay'` is assigned to `first_name` and `'Kulkarni'` to `last_name` due to their positions.

### Keyword Arguments

Keyword arguments are arguments passed to a function that are identified by their parameter name, rather than their position. This allows you to pass arguments in any order, improving readability and flexibility, especially for functions with many parameters or optional parameters.

**Benefits of Keyword Arguments:**
*   **Readability:** The code clearly indicates which value corresponds to which parameter.
*   **Flexibility:** The order of arguments doesn't matter.
*   **Easier with default values:** Can easily skip parameters with default values by not providing them, or provide them out of order.

Here's an example using keyword arguments for the `display_info` function:

```python
def display_info(first_name, last_name):
  print('First Name:', first_name)
  print('Last Name:', last_name)

# Calling the function using keyword arguments
display_info(last_name='Kulkarni', first_name='Akshay')
```

### Python Functions with Arbitrary Arguments (`*args`)

Sometimes, you might not know in advance how many arguments a function will receive. Python allows you to define functions that can take an arbitrary number of positional arguments using the `*args` syntax.

When `*args` is used as a parameter, it collects all positional arguments into a tuple.

```python
# Program to find the sum of multiple numbers
def find_sum(*numbers):
  result = 0
  print("Type of 'numbers' inside function:", type(numbers)) # It's a tuple
  for num in numbers:
    result = result + num
  print("Sum =", result)

# Function call with 3 arguments
print("Calling find_sum(5, 3, 7):")
find_sum(5, 3, 7)

# Function call with 2 arguments
print("\nCalling find_sum(24, 55):")
find_sum(24, 55)

# Function call with 0 arguments
print("\nCalling find_sum():")
find_sum()
```

## Python Variable Scope

Variable scope determines where a variable is accessible within a program. Python has different types of scope:

*   **Local Scope**: Variables defined inside a function are local to that function.
*   **Global Scope**: Variables defined outside any function are global.
*   **Nonlocal Scope**: Used in nested functions to refer to variables in an enclosing function's scope but not the global scope.

### Local Variables

A variable created inside a function belongs to the local scope of that function, and can only be accessed from within that function.

```python
# Local Variables example
def greet_local():
  message = 'Hello from local scope'
  print('Inside function (Local):', message)

greet_local()

# Attempting to access 'message' outside the function will cause an error
try:
  print('Outside function:', message)
except NameError as e:
  print('Error:', e, "- 'message' is a local variable and not accessible here.")
```

### Global Variables

A variable created in the main body of the Python script (outside any function) is a global variable. Global variables can be accessed from anywhere in the program, both inside and outside of functions.

```python
# Global Variables example
global_message = 'Hello from global scope'

def greet_global():
  # We can access global_message inside the function
  print('Inside function (Global):', global_message)

greet_global()

# We can also access global_message outside the function
print('Outside function (Global):', global_message)
```

### Nonlocal Variables

Nonlocal variables are used in nested functions (functions defined inside other functions). They allow an inner function to modify variables that are defined in its enclosing (outer) function's scope, but not in the global scope.

```python
# Python Nonlocal Variables example
def outer_function():
  a = 20 # 'a' is a non-local variable to inner_function
  print('Outer function (initial a):', a)

  def inner_function():
    nonlocal a # Declare 'a' as nonlocal to modify the 'a' in outer_function's scope
    a = a + 10
    print('Inner function (modified a):', a)

  inner_function()
  print('Outer function (final a):', a)

outer_function()
```

## Python Recursion

Recursion is a programming technique where a function calls itself in its definition. It's often used to solve problems that can be broken down into smaller, self-similar subproblems.

### Example: Factorial Calculation using Iteration

The factorial of a non-negative integer `n` is the product of all positive integers less than or equal to `n`. The base case for factorial is `0! = 1` and `1! = 1`.

```python
# Python Iterative example for Factorial
def factorial(x):
  """This is an iterative function to find the factorial of an integer"""
  if x < 0:
    return "Factorial does not exist for negative numbers"
  elif x == 0:
    return 1
  else:
    result = 1
    for i in range(1, x + 1):
      result *= i
    return result

num = 5
print("The factorial of", num, "is", factorial(num))
```
