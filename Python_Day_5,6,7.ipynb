### Tuple Methods: `count()` and `index()`

```python
a = (1, 2, 3, 4, 2, 5)
print(f"Original tuple 'a': {a}")

# Using count() to find occurrences of an element
count_of_2 = a.count(2)
print(f"Count of '2' in tuple 'a': {count_of_2}")

# Using index() to find the first occurrence of an element
index_of_3 = a.index(3)
print(f"Index of '3' in tuple 'a': {index_of_3}")

# You can specify a start and end index for index()
index_of_2_after_index_2 = a.index(2, 3) # Search for 2 starting from index 3
print(f"Index of '2' starting from index 3 in tuple 'a': {index_of_2_after_index_2}")
```

### Set Methods

```python
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
set3 = {7, 8}

print("Initial Set1: {set1}")

# add() - Adds an element to the set
set1.add(5)
print("After add(5): {set1}")

# update() - Adds elements from another iterable (e.g., list, tuple, set)
set1.update([6, 7])
print("After update([6, 7]): {set1}")

# union() - Returns a new set containing all items from both sets
union_set = set1.union(set2)
print("Union of Set1 and Set2: {union_set}")

# intersection() - Returns a new set containing only the items common to both sets
intersection_set = set1.intersection(set2)
print("Intersection of Set1 and Set2: {intersection_set}")

# difference() - Returns a new set containing items from the first set not in the second
difference_set = set1.difference(set2)
print("Difference (Set1 - Set2): {difference_set}")

# symmetric_difference() - Returns a new set with items present in either set, but not both
symmetric_difference_set = set1.symmetric_difference(set2)
print("Symmetric Difference: {symmetric_difference_set}")

# issubset() - Checks if all items in the set exist in the specified set
print("Is {{1, 2}} a subset of Set1? {{1, 2}}.issubset(set1): {{1, 2}}.issubset(set1)")

# issuperset() - Checks if all items in the specified set exist in the original set
print("Is Set1 a superset of {{1, 2}}? set1.issuperset({{1, 2}}): {set1.issuperset({1, 2})}")

# isdisjoint() - Checks if two sets have no items in common
print("Are Set1 and Set3 disjoint? set1.isdisjoint(set3): {set1.isdisjoint(set3)}")

# discard() - Removes the specified item (no error if not found)
set1.discard(1)
print("After discard(1): {set1}")
set1.discard(100) # Does nothing, no error
print("After discard(100): {set1}")

# clear() - Removes all elements from the set
set1.clear()
print("After clear(): {set1}")
```

## Dictionary

### Interview Question: Set vs. Dictionary
*   **Set:** A collection of unique, unordered elements.
*   **Dictionary:** A collection of key-value pairs, where keys must be unique and immutable.

```python
b={1:100,2:200,3:300}
b
```

```python
c={'name':'python','age':36,2:22.2}
c
```

### Features of a Dictionary
*   Duplicate keys are not allowed. If duplicate keys are provided during creation, the last value associated with that key will overwrite previous ones.

```python
g={1:10,1:1000,1:'java'}
g
```

### Key Characteristics of Dictionaries
*   **Unique Keys:** If multiple values are associated with the same key, the dictionary will retain only the last assigned value for that key.
*   **Iterable:** Dictionaries are iterable, allowing access to keys, values, or key-value pairs.
*   **Immutable Keys:** Dictionary keys must be immutable objects (e.g., numbers, strings, tuples). Mutable objects (like lists or sets) cannot be used as keys.
*   **Mutable Values:** Dictionary values can be any type of object, including mutable ones.

### Notes on Dictionary Methods
*   The `update()` method is used to add or modify key-value pairs in a dictionary.
*   If a key already exists, `update()` will change its corresponding value.
*   If a key does not exist, `update()` will add the new key-value pair.

### Dictionary Methods

```python
my_dict = {'name': 'Akshay', 'age': 23, 'city': 'Delhi'}
print("Initial Dictionary: {my_dict}")

# keys() - Returns a view object that displays a list of all the keys in the dictionary
print("Keys: {my_dict.keys()}")

# values() - Returns a view object that displays a list of all the values in the dictionary
print("Values: {my_dict.values()}")

# items() - Returns a view object that displays a list of a dictionary's key-value tuple pairs
print("Items: {my_dict.items()}")

# get() - Returns the value of the specified key. If the key does not exist, it returns None or a default value.
print("Value for 'name': {my_dict.get('name')}")
print("Value for 'country' (default None): {my_dict.get('country')}")
print("Value for 'country' (default 'USA'): {my_dict.get('country', 'USA')}")

# update() - Updates the dictionary with the specified key-value pairs
my_dict.update({'age': 23, 'occupation': 'Engineer'})
print("After update (existing key updated, new key added): {my_dict}")

# pop() - Removes the item with the specified key and returns its value
removed_city = my_dict.pop('city')
print("Removed city: {removed_city}, Dictionary after pop: {my_dict}")

# popitem() - Removes the last inserted key-value pair
# (in versions before 3.7, it removes an arbitrary item)
removed_item = my_dict.popitem()
print("Removed item: {removed_item}, Dictionary after popitem: {my_dict}")

# setdefault() - Returns the value of the specified key. If the key does not exist:
# inserts the key with the specified value, or None if no value is specified.
occupation_value = my_dict.setdefault('occupation', 'Developer') # Key 'occupation' exists, returns its value
print("setdefault 'occupation': {occupation_value}, Dictionary: {my_dict}")
country_value = my_dict.setdefault('country', 'India') # Key 'country' does not exist, adds it
print("setdefault 'country': {country_value}, Dictionary: {my_dict}")

# clear() - Removes all the elements from the dictionary
my_dict.clear()
print("After clear(): {my_dict}")
```
