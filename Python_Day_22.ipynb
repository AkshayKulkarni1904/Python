####Q.What is the diffrence between Series and DataFrame??(Interview Question)


A **Series** is a one-dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.). It's essentially a single column of data.

A **DataFrame** is a two-dimensional labeled data structure with columns of potentially different types. You can think of it like a spreadsheet or SQL table, or a dictionary of Series objects. It is the most commonly used pandas object.

```python
import pandas as pd

data=[120,30,50,45,11]
df=pd.Series(data)
df
```

Here, a Series `s` is created from a Python list `data`.

```python
#lets creat DF
df=pd.DataFrame([[10,20,30],[30,40,50]])
df
```

This creates a DataFrame `df` from a list of lists. Each inner list becomes a row in the DataFrame.

```python
df1=pd.DataFrame([[10,20,30],[30,40,50]],columns=['A','B','C'])
df1
```

You can also create a DataFrame and assign custom column names directly during creation, as shown above. This makes the data more readable and manageable.

```python
#Converst DataFrame to array using numpy
df.to_numpy()
```

The `.to_numpy()` method converts a DataFrame into a NumPy array. This can be useful when you need to perform numerical operations that are optimized for NumPy arrays, or when an external library expects a NumPy array as input.

###**Missing Values**

- `df.isna().sum()`: This counts the number of missing (NaN) values for each column.

However, choosing the right imputation strategy depends on the nature of the data and the problem.

###**Duplicate Data**

- `df.duplicated().sum()`: This method returns a boolean Series indicating whether each row is a duplicate of a previous row. Summing this Series gives the total number of duplicate rows.
- `df.drop_duplicates()`: This method removes duplicate rows from the DataFrame. By default, it keeps the first occurrence of a duplicate set. Similar to `dropna()`, you can use `inplace=True` to modify the DataFrame permanently.

```python
#Find duplicate rows
df.duplicated().sum()
```

```python
#Remove duplicates
df.drop_duplicates()
```

###**Filtering**

Filtering allows you to select rows based on one or more conditions, enabling you to focus on specific subsets of your data. This is a powerful feature for data exploration and analysis.

- **Single Condition**: `df[df["COLUMN"] > value]` selects rows where the specified column meets a single criterion.
- **Multiple Conditions (AND)**: `df[(condition1) & (condition2)]` combines conditions using the `&` operator, requiring all conditions to be true.
- **Multiple Conditions (OR)**: `df[(condition1) | (condition2)]` combines conditions using the `|` operator, requiring at least one condition to be true.
- **`between()` Method**: `df["COLUMN"].between(lower_bound, upper_bound)` selects rows where a numerical column falls within a specified range.
- **`isin()` Method**: `df["COLUMN"].isin(['value1', 'value2'])` selects rows where a column's value is present in a given list of values.

```python
#Players older than 30
df[df["AGE"]>30]
```

```python
#Overall rating above 90
df[df["OVERALL"]>90]
```

```python
#Left-footed players
df[df["PREFERRED_FOOT"]=='Left']
```

```python
#Right-footed players
df[df["PREFERRED_FOOT"]=="Right"]
```

```python
#Players aged between 20 and 25
df[df["AGE"].between(20,25)]
```

```python
#Barcelona players
df[df["CLUB"]=="FC Barcelona"]
```

```python
#Age >30 and Overall >85
df[(df["AGE"]>30)&(df["OVERALL"]>85)]
```

```python
#Brazil or Argentina players
df[df['NATIONALITY'].isin(['Brazil','Argentina'])]
```

###**GroupBy**

The `groupby()` method is used to split the DataFrame into groups based on some criteria. Once grouped, you can apply an aggregation function (like `mean()`, `sum()`, `count()`) to each group. This is a powerful tool for analyzing relationships and patterns across different categories.

- `df.groupby("GROUPING_COLUMN")["TARGET_COLUMN"].mean()`: Groups the DataFrame by `GROUPING_COLUMN` and then calculates the mean of `TARGET_COLUMN` for each group.
- This can be extended to `count()`, `sum()`, `max()`, `min()`, etc., to get different aggregated statistics for each group.

```python
#Average age by nationality
df.groupby("NATIONALITY")["AGE"].mean()
```

```python
#Average overall by position
df.groupby("POSITION")["OVERALL"].mean()
```

```python
#Number of players in each club
df.groupby("CLUB")["NAME"].count()
```

```python
#Average potential by preferred foot
df.groupby("PREFERRED_FOOT")["POTENTIAL"].mean()
```

####**Top Players**

Identifying top or bottom performers or values in a dataset is a common analytical task. Pandas provides convenient methods for this.

- `df.nlargest(n, "COLUMN")`: Returns the `n` rows with the largest values in the specified column.
- `df.nsmallest(n, "COLUMN")`: Returns the `n` rows with the smallest values in the specified column.
- `df.loc[df["COLUMN"].idxmax()]`: Returns the entire row corresponding to the maximum value in the specified column.
- `df.loc[df["COLUMN"].idxmin()]`: Returns the entire row corresponding to the minimum value in the specified column.

```python
#Top 10 players by Overall
df.nlargest(10,"OVERALL")
```

```python
#Top 5 youngest players
df.nsmallest(5,"AGE")
```

```python
#Player with highest Overall
df.loc[df["OVERALL"].idxmax()]
```

```python
#Player with highest Potential
df.loc[df["POTENTIAL"].idxmax()]
```
