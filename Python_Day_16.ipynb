## Solution for Assignment: Count occurrences of each character in a string

Here are a few ways to solve this assignment:


````python
# Method 1: Using a dictionary and a loop
s = 'simple dimple'
char_counts = {}
for char in s:
    char_counts[char] = char_counts.get(char, 0) + 1
print(char_counts)
````

````python
# Method 2: Using collections.Counter
from collections import Counter
s = 'simple dimple'
char_counts_counter = Counter(s)
print(char_counts_counter)
````

````python
# Method 3: Using dictionary comprehension (less readable for counting)
# This is more suited for creating a dict based on existing values, not directly counting
# However, it can be combined with count() method for a single character
s = 'simple dimple'
char_counts_comp = {char: s.count(char) for char in s}
print(char_counts_comp)
````

## Understanding the `filter()` function

The `filter()` function in Python is a built-in higher-order function that takes two arguments:
1.  **A function (or `None`):** This function is applied to each item of an iterable. It should return `True` or `False`.
2.  **An iterable:** This is the sequence (like a list, tuple, or string) whose elements will be filtered.

### How it works:

`filter()` constructs an iterator from those elements of `iterable` for which `function` returns `True`. In essence, it 'filters out' elements that do not satisfy the condition defined by the function.

### Key characteristics:

*   **Conditional selection:** It's primarily used for conditional selection of elements from an iterable.
*   **Returns an iterator:** Like `map()`, `filter()` returns an iterator (a `filter` object) and not directly a list or tuple. You typically need to convert it (e.g., using `list()` or `tuple()`) to see the filtered elements.
*   **`None` as function:** If the function argument is `None`, `filter()` will remove elements that are considered 'falsy' (e.g., `0`, `False`, `None`, `''`, `[]`, `{}`).

**Example:**
```python
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers) # Output: [2, 4, 6]
```

The `filter()` function constructs an iterator from elements of an iterable for which a function returns true. It's used for conditionally selecting elements. Here, it filters out all negative numbers from the tuple `p`.

````python
list(filter(lambda num:num>0,p))
````

````python
#Write a function to check whether a number is prime
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):  # check up to sqrt(n)
        if n % i == 0:
            return False
    return True
is_prime(25)
````

````python
#Write a function to find the factorial of a number.
def fact(n):
  if n<0:
    return "Factorial not defined for negative numbers"
  elif n==0 or n==1:
    return 1
  else:
    result =1
    for i in range(2,n+1):
      result *= i
    return result
fact(5)
````
