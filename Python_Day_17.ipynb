**Q. map vs filter**

Both `map()` and `filter()` are built-in functions in Python that operate on iterables (like lists, tuples, etc.) and return an iterator (which can be converted to a list, tuple, etc.).

*   **`map(function, iterable)`:**
    *   Applies a given `function` to each item of an `iterable`.
    *   Transforms each item based on the function's logic.
    *   The result is a new iterable containing the *transformed* items.
    *   The length of the output iterable is always the same as the input iterable.
    *   Example: `list(map(lambda x: x*2, [1, 2, 3]))` would result in `[2, 4, 6]`.

*   **`filter(function, iterable)`:**
    *   Constructs an iterator from elements of an `iterable` for which a `function` returns true.
    *   Selects items from the iterable based on a condition (the function must return a boolean).
    *   The result is a new iterable containing only the *filtered* items.
    *   The length of the output iterable can be less than or equal to the input iterable.
    *   Example: `list(filter(lambda x: x > 2, [1, 2, 3, 4]))` would result in `[3, 4]`.

**In summary:** `map` transforms elements, while `filter` selects elements.

**Q. What do you mean by python is interpreted language**

When we say Python is an "interpreted language," it means that Python code is executed line by line by an interpreter, rather than being fully compiled into machine-readable code (like C++ or Java) before execution.

Here's a breakdown:

1.  **Source Code to Bytecode:** When you run a Python script, the Python interpreter first compiles the human-readable source code (`.py` files) into an intermediate format called **bytecode** (`.pyc` files). This bytecode is a low-level, platform-independent representation of your code.
2.  **Bytecode Execution:** This bytecode is then executed by the Python **Virtual Machine (PVM)**, which is the runtime engine of Python. The PVM interprets the bytecode instructions and translates them into machine code that the computer's CPU can understand and execute.
3.  **No Separate Compilation Step:** The key difference from a compiled language is that this compilation to bytecode happens *automatically* and *just-in-time* when you run the script, often without the user explicitly invoking a separate compiler. The execution starts almost immediately, rather than waiting for a potentially long compilation process to finish.

**Advantages of interpreted languages like Python:**
*   **Platform Independence:** Bytecode can run on any platform that has a Python interpreter, making Python highly portable.
*   **Faster Development Cycle:** No explicit compilation step means developers can quickly test and run code changes.
*   **Dynamic Features:** Easier to implement dynamic features like dynamic typing and reflection.

**Disadvantages:**
*   **Slower Execution:** Generally, interpreted code runs slower than fully compiled code because of the overhead of the interpreter translating bytecode to machine code at runtime.

### Understanding `reduce()`

The `reduce()` function is part of the `functools` module in Python. It applies a given function to the items of an iterable in a cumulative way, reducing the iterable to a single output value.

It requires two arguments: a function (usually a `lambda` function) that takes two parameters, and an iterable (like a list). The function is first applied to the first two items, then to the result of that operation and the third item, and so on, until all items are processed. For example, `reduce(lambda x,y:x+y, [1,2,3,4])` would calculate `((1+2)+3)+4`.

```python
#3. reduce(function,iterable)---> it is used to convert collection into a single value/object
#reduce is present in functools module
# reduce works on 2 args no more no less

from functools import reduce

k=[2,3,4,5,6]
reduce(lambda x,y:x+y,k) #in reduce we dont need typecasting as it gives the result
```

```python
reduce(lambda x,y:max(x,y),k)
```

### Python Modules: Organization and Reusability

In Python, a **module** is simply a file containing Python definitions and statements. The filename is the module name with the `.py` extension. Modules allow you to logically organize your Python code, making it reusable and easier to manage. They can contain:

*   **Variables:** Data stored under a name.
*   **Functions:** Reusable blocks of code.
*   **Classes:** Blueprints for creating objects.
*   **Constants:** Variables whose values are not intended to change.

There are generally two types of modules:

1.  **Built-in Modules:** These are modules that come pre-installed with Python (e.g., `math`, `os`, `sys`, `functools`).
2.  **User-defined Modules:** These are modules created by developers to encapsulate their own functions, classes, and variables.

### Importing Modules

To use any module (built-in or user-defined), you first need to import it into your current script or notebook. The most basic syntax for importing a module is:

`import module_name`

Once imported, you can access its members (functions, variables, etc.) using the dot notation: `module_name.member_name`.

```python
# import math module
import math
```

Here, we import the `math` module and then demonstrate accessing some of its built-in constants (`math.e` for Euler's number, `math.pi` for Pi) and functions (`math.factorial()` for calculating factorials).

```python
math.e #eulars constant
```

```python
math.pi
```

```python
math.factorial(5)
```

### Module Aliasing: Shorter Names for Modules

When a module has a long name, or to avoid naming conflicts with other modules or variables in your code, you can assign it a shorter alias (or nickname) during the import process. The syntax is:

`import module_name as alias_name`

After aliasing, you can use the `alias_name` to refer to the module and access its members (e.g., `alias_name.member_name`).

```python
m.sqrt(25)
```

```python
m.pow(2,3)
```

```python
#To import only specific and required things we need to go with
#from module_name import member
from math import pi,factorial
```

### Importing Specific Members from a Module

Instead of importing the entire module, you can choose to import only specific functions, variables, or classes you need. This can make your code cleaner and avoid namespace clutter. The syntax is:

`from module_name import member1, member2, ...`

When you import members this way, you can use them directly by their name without needing to prefix them with the module name (e.g., `pi` instead of `math.pi`).

```python
pi
```

```python
factorial(5)
```

```python
#member aliasing - assign a short name to the member of module
from math import factorial as f, sqrt as s
```

### Member Aliasing: Renaming Imported Members

Just like modules, individual members imported from a module can also be given aliases. This is useful for shortening long member names or resolving name conflicts if you import multiple members with the same name from different modules. The syntax is:

`from module_name import original_member_name as alias_name`

Now, you can use `alias_name` directly in your code.

```python
f(4)
```

```python
s(25)
```

```python
#to import all members in 1 go use * approch
from math import * #(to access all members of a module directly we use *)
```

### Wildcard Imports: Importing All Members

The `from module_name import *` syntax imports *all* members (functions, variables, classes) from a module directly into the current namespace. This means you can use any of the module's members without any prefix.

**Caution:** While convenient, wildcard imports are generally discouraged in larger projects because they can lead to namespace collisions (where two different modules might define members with the same name), making it harder to track where a particular function or variable came from. It's best used sparingly, often in interactive sessions or small scripts where clarity isn't compromised.

```python
from bank_info import *
```

### Solution to Practice Questions

#### Question 1: Reduce Function

```python
from functools import reduce

numbers = [10, 20, 30, 40]
product = reduce(lambda x, y: x * y, numbers)
print("The product of the list {numbers} is:", {product})
```

#### Question 2: Math Module

```python
import math

# Square root of 144
sqrt_144 = math.sqrt(144)
print("Square root of 144:", {sqrt_144})

# 3 raised to the power of 4
pow_3_4 = math.pow(3, 4)
print("3 raised to the power of 4:", {pow_3_4})

# Factorial of 7
factorial_7 = math.factorial(7)
print("Factorial of 7:", {factorial_7})
```

#### Question 3: Module Aliasing

```python
import math as mt

ceiling_3_14 = mt.ceil(3.14)
print("Ceiling of 3.14 using alias 'mt':" {ceiling_3_14})
```

#### Question 4: Specific Imports & Member Aliasing

```python
from math import log as ln, e

log_e = ln(e)
print("Natural logarithm of e using alias 'ln':", {log_e})
```

#### Question 5: Wildcard Import

```python
from math import *

sin_0 = sin(0)
cos_0 = cos(0)
print("sin(0):",{sin_0})
print("cos(0):", {cos_0})
```
